```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

scatter1 = axes[0].scatter(df1[df1.columns[1]], df1[df1.columns[2]], c=df1['Life satisfaction'], cmap='twilight', s=df1['GDP per capita (USD)']/1000, label='Raw data')
axes[0].set_xlabel(df1.columns[1])
axes[0].set_ylabel(df1.columns[2])
axes[0].set_title('Life Satisfaction')
cbar1 = plt.colorbar(scatter1, ax=axes[0])
axes[0].plot([25000, 60000], [5.44377435, 7.81638924], color='red', linewidth=2, label='Linear model')
axes[0].legend()

scatter2 = axes[1].scatter(df2[df2.columns[1]], df2[df2.columns[2]], c=df2['population'], cmap='inferno', s=df2['day']/100, label='Raw data')
axes[1].set_xlabel(df2.columns[1])
axes[1].set_ylabel(df2.columns[2])
cbar2 = plt.colorbar(scatter2, ax=axes[1])
axes[1].plot([520, 520], [0, 155000], color='green', linestyle='--', linewidth=2, label='Key date')
axes[1].legend()

plt.show()
```
